```python
import pandas as pd

# This section displays all 12 columns
pd.set_option('display.max_columns',None)
# This section is used to load the data using Pandas and display the data
dataFrame = pd.read_csv("healthcare-dataset-stroke-data.csv")
display(dataFrame)
print(dataFrame)

#This section prints total number of missing or null data in each columns
print(dataFrame.isnull().sum())
#This section is used to display only the rows with missing or null data
missing_rows = dataFrame[dataFrame.isnull().any(axis=1)]
display(missing_rows)

#This section is part of the cleaning steps where the rows with missing or null data was removed since the total number of rows with missing data were 201 and with the large dataset, removing 201 data rows will not cause big diffrence.
dataFrame= dataFrame.dropna()

#This section was used to find another missing data regarding the smoking status of patients labeled as unknown.
smoking_rows = dataFrame[dataFrame['smoking_status']=='Unknown']
#This section is used to display the rows with unknown smoking status and get the total number of rows in the dataset
display(smoking_rows)
count_smoke= dataFrame[dataFrame['smoking_status']=='Unknown'].shape[0]
print("Number of Unknown Smokers:", count_smoke)

#To futhur clean the data, any rows with 2 or more missing data was drop or removed from the data set. In this section, variables with null value in bmi column and variable with unknown in smoking status was removed. The data was removed since the data set size was large and the removal of few data with missing values caused the data to be more accurate.
dataFrame = dataFrame[~((dataFrame['bmi'].isnull()) & (dataFrame['smoking_status'] == 'Unknown'))]
count_both= dataFrame[~((dataFrame['bmi'].isnull()) & (dataFrame['smoking_status'] == 'Unknown'))].shape[0]
display(dataFrame)

#This section was used to find and drop any duplicates data present in the dataset
dataFrame = dataFrame.drop_duplicates()

#This dataset consists the value of 0 and 1 for No and Yes in certain column which can be confusing as well as be seen as noisy data. To address this issue, the 0 and 1 on the three column, hypertension, heart diseease and stroke was changed to No and Yes for easier understanding and smooth data
dataFrame[['hypertension','heart_disease','stroke']] = dataFrame[['hypertension','heart_disease','stroke']].replace({0: 'No', 1: 'Yes'})
display(dataFrame)

#This section is used to conduct exploratory data analysis using visualizations (Matplotlib)
import matplotlib.pyplot as plt

#This section is used to describe a quick summary stats of the data set
dataFrame.describe()

#This section is used to find the data size
dataFrame.shape


#This section was used to conduct the EDA process and see various visualization to find patterns and key insights.
#This section was used to determine the noisy or unwanted data as well as outliers and remove then for a clean and smooth data.
#In this section IQR was performed to determine the lower and upper limit to filter out the data.
#With finding the IQR and outlier, the data that were out of place or not within range were filtered out for various columns.
bmi_IQR = dataFrame['bmi'].quantile(0.75) - dataFrame['bmi'].quantile(0.25)
bmi_low_limit = dataFrame['bmi'].quantile(0.25) - 1.5 * bmi_IQR
bmi_upper_limit = dataFrame['bmi'].quantile(0.75) + 1.5 * bmi_IQR
dataFrame[dataFrame['bmi'] > bmi_upper_limit]
dataFrame.drop(dataFrame[dataFrame['bmi'] < bmi_low_limit].index, inplace = True)
dataFrame.drop(dataFrame[dataFrame['bmi'] > bmi_upper_limit].index, inplace = True)
#Box Plot and Histogram was plotted to see the pattern and analysis the data set
plt.figure()
plt.boxplot(dataFrame['bmi'])
plt.figure()
dataFrame['bmi'].hist()


age_IQR = dataFrame['age'].quantile(0.75) - dataFrame['age'].quantile(0.25)
age_low_limit = dataFrame['age'].quantile(0.25) - 1.5 * age_IQR
age_upper_limit = dataFrame['age'].quantile(0.75) + 1.5 * age_IQR
dataFrame[dataFrame['age'] > age_upper_limit]
dataFrame.drop(dataFrame[dataFrame['age'] < age_low_limit].index, inplace = True)
dataFrame.drop(dataFrame[dataFrame['age'] > age_upper_limit].index, inplace = True)
plt.figure()
plt.boxplot(dataFrame['age'])
plt.figure()
dataFrame['age'].hist()


avg_glucose_level_IQR = dataFrame['avg_glucose_level'].quantile(0.75) - dataFrame['avg_glucose_level'].quantile(0.25)
avg_glucose_level_low_limit = dataFrame['avg_glucose_level'].quantile(0.25) - 1.5 * avg_glucose_level_IQR
avg_glucose_level_upper_limit = dataFrame['avg_glucose_level'].quantile(0.75) + 1.5 * avg_glucose_level_IQR
dataFrame[dataFrame['avg_glucose_level'] > avg_glucose_level_upper_limit]
dataFrame.drop(dataFrame[dataFrame['avg_glucose_level'] < avg_glucose_level_low_limit].index, inplace = True)
dataFrame.drop(dataFrame[dataFrame['avg_glucose_level'] > avg_glucose_level_upper_limit].index, inplace = True)
plt.figure()
plt.boxplot(dataFrame['avg_glucose_level'])
plt.figure()
dataFrame['avg_glucose_level'].hist()


```
